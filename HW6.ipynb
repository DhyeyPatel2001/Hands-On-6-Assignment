```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[len(arr) // 2]  # Non-random pivot: middle element
        left = [x for x in arr if x < pivot]
        middle = [x for x in arr if x == pivot]
        right = [x for x in arr if x > pivot]
        return quicksort(left) + middle + quicksort(right)
```

```python
import random

def quicksort_random(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = random.choice(arr)  # Random pivot
        left = [x for x in arr if x < pivot]
        middle = [x for x in arr if x == pivot]
        right = [x for x in arr if x > pivot]
        return quicksort_random(left) + middle + quicksort_random(right)
```

```python
import time
import matplotlib.pyplot as plt
import numpy as np

# Best case: Already sorted array
def generate_best_case(n):
    return list(range(n))

# Worst case: Reverse sorted array
def generate_worst_case(n):
    return list(range(n, 0, -1))

# Average case: Random array
def generate_average_case(n):
    return np.random.randint(0, 1000, size=n).tolist()

# Benchmark function
def benchmark_quicksort(sizes, case_type):
    times = []
    for n in sizes:
        if case_type == 'best':
            arr = generate_best_case(n)
        elif case_type == 'worst':
            arr = generate_worst_case(n)
        elif case_type == 'average':
            arr = generate_average_case(n)
        
        start_time = time.time()
        quicksort(arr)
        times.append(time.time() - start_time)
    return times

# Define the sizes of arrays to test
sizes = [100, 500, 1000, 5000, 10000]

# Measure time for each case
best_times = benchmark_quicksort(sizes, 'best')
worst_times = benchmark_quicksort(sizes, 'worst')
average_times = benchmark_quicksort(sizes, 'average')

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(sizes, best_times, label="Best Case", marker='o')
plt.plot(sizes, worst_times, label="Worst Case", marker='o')
plt.plot(sizes, average_times, label="Average Case", marker='o')
plt.xlabel("Input size (n)")
plt.ylabel("Execution time (seconds)")
plt.title("Quicksort Performance for Best, Worst, and Average Cases")
plt.legend()
plt.grid(True)
plt.show()
```
